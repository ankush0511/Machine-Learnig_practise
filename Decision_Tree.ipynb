```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.data
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
##independent feature
x=df.iloc[:,:-1]
y=iris.target
```

```python
## tarn test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
# post pruning technique (all default is selected)
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
##decisiton tree mai hyper parameter
treemodel1=DecisionTreeClassifier(max_depth=2,)
```

```python
treemodel1.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,5))
tree.plot_tree(treemodel1,filled=True)
```

```python
#prediction
y_pred=treemodel1.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
score=accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

##preprunning

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    'ccp_alpha':[1,2,3,4,5,6,7]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel=DecisionTreeClassifier()
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
cv.predict(x_test)
#it gives all 1 it is overfitting (so we remove the ccp_alpha)
```

```python
y_pred=cv.predict(x_test)
```

```python
score=accuracy_score(y_pred,y_test)
score
```
