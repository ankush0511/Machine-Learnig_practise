### with pipeline

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
#missing value->OHE -> scaling ->feature selection -> decision Tree
df.drop(columns=['Cabin','Ticket','Name','PassengerId'],inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),
                                              df['Survived'],
                                              test_size=0.2,
                                              random_state=42)
```

```python
x_train.head()
```

```python
#imputatin transform
trf1=ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
#OHE
trf2=ColumnTransformer([
    ('ohe_s_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
#scaling
trf3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
#feature selection
trf4=SelectKBest(score_func=chi2,k=5)
```

```python
#train the model
trf5=DecisionTreeClassifier()
```

### create Pipeline

```python
pipe=Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

```python
# alternate
pipe=make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.named_steps
```

```python
pipe.named_steps['trf1'].transformers_
```

```python
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

## crossvalidation using pipeline

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
cross_val_score(pipe,x_train,y_train,cv=5,scoring='accuracy').mean()
```

## Grid search using pipeline

```python
#trf5 is a model name
parms={
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
grid=GridSearchCV(pipe,parms,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
grid.best_score_
```
